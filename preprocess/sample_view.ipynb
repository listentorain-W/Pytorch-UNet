```python
"""
Author      : Tingyu Wang
Description : For preprocess Chinese Sea Surface Temperature index prediction
Version     : 0.1
Changelog   : Change preprocess region to South China Sea
"""

import re
import warnings
import time
import numpy as np
import xarray as xr
import pathlib as pl
from eofs.xarray import Eof
from scipy import signal

def mask(ds, label='land'):
    landsea = xr.open_dataset("../../temp_data/mask/landsea.nc")
    landsea = landsea['LSMASK']

    # ds和地形数据分辨率不一致，需将地形数据插值
    # fill_value 使0经度的NaN变为数值，防止出现白边
    landsea = landsea.interp(
                            lat=ds.lat.values, 
                            lon=ds.lon.values,
                            kwargs={'fill_value': None}
                             )

    # 利用地形掩盖海陆数据
    if label == 'ocean':                  # when cond is False
        ds = ds.where(landsea.data == 0)  # default locations fill with NaN
    elif label == 'land':
        ds = ds.where(landsea.data == 1)
    elif label == 'lake':
        ds = ds.where(landsea.data == 2)
    elif label == 'small island':
        ds = ds.where(landsea.data == 3)
    elif label == 'ice shelf':
        ds = ds.where(landsea.data == 4)
    return ds

def rm_annual_cycle(ds):
    ds = ds.groupby('time.month') - ds.groupby('time.month').mean()
    ds = ds.reset_coords('month', drop = True)
    return ds


def data_preprocess(ds, lat_range = 0, lon_range = 0, time_range = 0, calculate_type = None):
    """
    slice region
    """
    beginning = ds.time[0]
    ending = ds.time[-1]
    bgn_yr, end_yr = int(beginning / 100), int(ending / 100)
    bgn_mon, end_mon = int(beginning - (bgn_yr * 100)), int(ending - (end_yr * 100))
    start_ym = str(bgn_yr).zfill(4) + '-' + str(bgn_mon).zfill(2)
    end_ym = str(end_yr).zfill(4) + '-' + str(end_mon).zfill(2)
    ds['time'] = xr.cftime_range(
                    start = start_ym, 
                    end = end_ym,        # Changelog: change period to end
                    freq = 'MS',
                    calendar = 'noleap'
                    )

    # calculate the year mean
    # ds_end = ds.groupby('time.year').mean()
    # ds_end = ds_end.rename({'year': 'time'})
    ds_end = ds
    return ds_end # 一生的耻辱 原本写成了ds，导致白算了

def pick_nino34(SSTA):
    nino34 = SSTA.sel(lat = slice(-5, 5), lon = slice(170, 240)).mean(dim =['lat','lon'])
    return nino34

def pick_AMO(SSTA):
    # SSTA_detrend = signal.detrend(SSTA, axis = 0, type = 'linear')
    AMO = SSTA.sel(lat = slice(0, 65), lon = slice(280, 360)).mean(dim = ['lat', 'lon'])
    AMO_detrend = signal.detrend(AMO, axis = 0, type = 'linear')
    return AMO_detrend

def pick_PDO(SSTA):
    # to get leading PC1 ==> PDO index
    ssta_pdo = SSTA - SSTA.mean(dim = ['lat', 'lon'])
    ssta_pdo_ = ssta_pdo.sel(lat = slice(20, 70), lon = slice(110, 260))
    coslat = np.cos(np.deg2rad(ssta_pdo_.coords['lat'].values))
    wgts = np.sqrt(coslat)[..., np.newaxis]
    solver = Eof(ssta_pdo_, weights=wgts)

    # no need to calculate EOF pattern
    # eof1 = solver.eofsAsCorrelation(neofs=1)
    pc1 = solver.pcs(npcs=1, pcscaling=1)

    # leading PC
    PDO = pc1[:, 0]
    return PDO

def pick_CHN(SSTA, region):
    region = 'South China Sea'
    if region == 'South China Sea':
        latS, latN = 2, 22
        lonW, lonE = 105, 120
    print('you are in Nan Hai')
    index = SSTA.sel(lat = slice(latS, latN), lon = slice(lonW, lonE)).mean(dim = ['lat', 'lon'])
    return index

def index_calcu(SSTA, y_name):
    # Nino34
    if y_name == 'nino34-index':
        pc = pick_nino34(SSTA)

    # PDO index
    elif y_name == 'PDO-index':
        pc = pick_PDO(SSTA)

    # AMO index
    elif y_name == 'AMO-index':
        pc = pick_AMO(SSTA)
    
    # CHN Ocean SST index
    elif y_name in ['South China Sea', 'East China Sea']:
        pc = pick_CHN(SSTA, y_name)
        
    return pc

def Normalization(X):
    return ((X - np.mean(X)) / (np.max(X) - np.min(X)))

def get_layer(ds, top = -1, bottom = -1 ):
    ds = ds.loc[:, top: bottom, ...]  
    layer = ds.mean(axis = 1)
    # ds = ds.sel(lev = slice(top, bottom))
    # layer = ds.mean(dim = ['lev'])
    return layer
    
    
def get_tendency(ds, var_name = 'ts', ds_path = ' '):
    bgning = 195011.
    ending = 201410.
    tend_ds = xr.open_dataset(ds_path)[var_name]\
                .sel(time = slice(bgning, ending))
    if var_name == 'ts':
        tend_ds = tend_ds.sel(lat = slice(-30, 55), lon = slice(80, 160))
        tend_ds = mask(tend_ds, 'ocean')
        tend_ds = data_preprocess(tend_ds)
        tend_ano = rm_annual_cycle(tend_ds)
        
    elif var_name == 'thetao':
        tend_ds = tend_ds.sel(lat = slice(-30, 55), lon = slice(80, 160))
        tend_ds = data_preprocess(tend_ds)
        tend_ds = get_layer(tend_ds, top = 0, bottom = 300)
        tend_ano = rm_annual_cycle(tend_ds)
        
    tendency = ds - tend_ano.values
    return tendency

def nc_encoding():
    # dataset
    # convert Dataset to netcdf file named 'train data' or 'dev data'
    encoding = {'lat': {
                        '_FillValue': 1e+20,
                        'dtype': 'float32',
                        'zlib': False
                        },
                'lon': {
                        '_FillValue': 1e+20,
                        'dtype': 'float32',
                        'zlib': False
                        },
                'channel': {
                        'dtype': 'int32',
                        'zlib': False
                        },
                'lead_time': {
                        'dtype': 'int32',
                        'zlib': False
                        },
                'sample_num': {
                            'dtype': 'int32',
                            'zlib': False
                            },
                'X':{
                    '_FillValue': 1e+20,
                    'zlib':False
                    },
                'y':{
                    '_FillValue': 1e+20,
                    'zlib':False
                    }
            }
    return encoding


if __name__ == "__main__":
    warnings.filterwarnings('ignore')
    # ========================= Set parameters =========================
    
    # 原本batch_size = 62, bgning = 195001.
    # 为了使用tendency故减少一年
    batch_size = 61
    batch_num = 12
    bgning, ending = 195101., 201412.
    
    # y_name = 'PDO-index' or 'nino34-index' or 'AMO-index' or 'South China Sea'
    # X_name = 'detrend-SSTA' 'SSTA'
    X_name, y_name = 'SSTA', 'South China Sea' 
    X_len, y_len = 1, 24  # time length in X, y Unit: / yr
    layer_num = 3
    
    mini_yr = batch_size * batch_num + (X_len + y_len) - 1
    max_yr = 65 * 12
    
    hostname = 'test'  # 'OMEN' 'ZHAN' 'Group-Server'
    data_type = 'train'  # 'train' or 'dev'
    split_type = 'ALL'  # '12' or 'ALL'
    X_, y_ = [], []
    model_list = []
    filter_type = 'lowpass'  # 'lowpass', 'bandpass', 'no_filter'
    
    model_select = ['E3SM-1-1-ECA', 'E3SM-1-1', 'E3SM-1-0']
    
    if hostname == 'OMEN':
        ts_dir = '../../CMIP6/historical/Amon/ts/'
        thetao_dir = '../../CMIP6/historical/Omon/thetao/thetao_'
    elif hostname == 'ZHAN':
        ts_dir = '../../../CMIP6/historical/Amon/ts/'
        thetao_dir = '../../../CMIP6/historical/Omon/thetao/thetao_'
    elif hostname == 'Group-Server':
        ts_dir = '/disk2/CMIP6/preprocess/output/historical/Amon/ts/'
        thetao_dir = '/disk2/CMIP6/preprocess/output/historical/Omon/thetao/thetao_'
    elif hostname == 'test':
        ts_dir = '/disk1/tywang/CMIP6/test/ts/'
        thetao_dir = '/disk1/tywang/CMIP6/test/thetao/thetao_'  
    
    
    # get all CMIP6 piControl data path
    filelist = list(pl.Path(ts_dir).rglob("*.nc")) 

    # ========================= Process each model data =========================
    deal_num = 0
    for i, fpath in enumerate(filelist):
        # print(fpath.name)
        
        # read piControl data
        model_name = re.findall(r".+_(.+?)_.{29}", fpath.name)[0]
        if model_name in model_select:
            print(f"{model_name} will jump!\n")
            continue
        
        ts = xr.open_dataset(fpath)['ts']\
               .sel(time = slice(bgning, ending))

        # use try except to avoid no file problem
        thetao_path = thetao_dir + model_name + '_Hist_unigrid_195001-201412.nc'
        try:
            thetao = xr.open_dataset(thetao_path)['thetao']\
                       .sel(time = slice(bgning, ending))
        except OSError:
            print(f'xxxxxxxx Sorry, {model_name} thetao data does not exist. xxxxxxxx\n')
            continue
        
        # if ts.time[bgning] != thetao.time[bgning] :
            # raise SystemExit(f'ts_start = {ts.time[bgning]} \nthetao_start = {thetao.time[bgning]}\n' + 
            #                   '********start time is not equal!********')
        #    print(f'ts_start = {ts.time[bgning]} \nthetao_start = {thetao.time[bgning]}\n' + 
        #                       '-- Start time is not equal! --')
        #    continue
        # ts_tlen = len(ts.time) / 12
        # if ts_tlen < max_yr: # 499 是最大可利用数据数量，
        #    print(f'{model_name} is less than {max_yr}')
        #    continue                 # 同时也是保证能至少使用28个model
        
        print(f"******** Model name :{model_name} is processing... ********")
        model_list.append(model_name)

        # ========================= Slice data =========================

        # slice data 55S - 70N and time from begin to end
        # SSTA
        ts = ts.sel(lat = slice(-30, 55), lon = slice(80, 160))
        ts = mask(ts, 'ocean')
        ts = data_preprocess(ts)
        SSTA = rm_annual_cycle(ts)
        
        # Thetao (Multi-layer)
        thetao = thetao.sel(lat = slice(-30, 55), lon = slice(80, 160))
        thetao = data_preprocess(thetao)
        
        layer1 = get_layer(thetao, top = 0, bottom = 300)
        layer1_ano = rm_annual_cycle(layer1)
        
        # Tendency
        # layer2 = get_layer(thetao, top = 301, bottom = 700)
        # layer2_ano = rm_annual_cycle(layer2)
        layer2_ano = get_tendency(SSTA, var_name = 'ts', ds_path = fpath)
        
        # ========================= y calculate =========================
        
        pc = index_calcu(SSTA, y_name)
        
        # ========================= X calculate =========================
        # ========================= filter SSTA / ssta_pdo, index =========================
        
        # choose X-type
        if X_name == 'detrend-SSTA':
            SSTA = SSTA - SSTA.mean(dim = ['lat', 'lon'])
        
        # use Butterw filter
        order = 9
        if filter_type == 'lowpass':
            Wn = 1/3   # filter lower than 3 yr signal Wn = 1 / (year)
            b, a = signal.butter(order, Wn, filter_type)
            y_filt = signal.filtfilt(b, a, pc, axis = 0)
            SSTA_filt = signal.filtfilt(b, a, SSTA, axis = 0)
            ly1_filt = signal.filtfilt(b, a, layer1_ano, axis = 0)
            ly2_filt = signal.filtfilt(b, a, layer2_ano, axis = 0)
            filter_time = str(1 / Wn)
            
        elif filter_type == 'bandpass':
            Wn = [1/96, 1/3]
            b, a = signal.butter(order, Wn, filter_type)
            y_filt = signal.filtfilt(b, a, pc, axis = 0)
            SSTA_filt = signal.filtfilt(b, a, SSTA, axis = 0)
            ly1_filt = signal.filtfilt(b, a, layer1_ano, axis = 0)
            ly2_filt = signal.filtfilt(b, a, layer2_ano, axis = 0)
            filter_time = str(1/Wn[1]) +'-'+ str(1/Wn[0])
            
        elif filter_type == 'no_filter':
            Wn = 1
            y_filt = pc.values
            SSTA_filt = SSTA.values
            ly1_filt = layer1_ano.values
            filter_time = "no-filter"
        
        # Normalization
        if y_name == 'PDO-index':
            y_filt = Normalization(y_filt)

        # ========================= Reorganize data =========================
        
        # fill NaN with 0.
        SSTA_filt[np.isnan(SSTA_filt)] = 0
        surface_data = SSTA_filt
        surface_data = np.expand_dims(surface_data.astype(np.float32), axis = -1)
        
        ly1_filt[np.isnan(ly1_filt)] = 0
        layer1_data = ly1_filt
        layer1_data = np.expand_dims(layer1_data.astype(np.float32), axis = -1)
        
        ly2_filt[np.isnan(ly2_filt)] = 0
        layer2_data = ly2_filt
        layer2_data = np.expand_dims(layer2_data.astype(np.float32), axis = -1)
        
        label = y_filt
        label = label.astype(np.float32)
        
        print(f'surface shape: {surface_data.shape}')
        print(f'layer1 shape: {layer1_data.shape}')
        print(f'layer2 shape: {layer2_data.shape}')
        
        if data_type == 'train':
            loop_list = range(batch_size * batch_num)
        elif data_type == 'dev':                 # dev 数据的切片还需再确认一次
            loop_list = range(mini_yr, 
                              max_yr - X_len - y_len + 1)

        # 切片索引[start: end], end的位置是取不到的
        for m in loop_list:
            X_store = np.zeros([1, len(SSTA.lat), len(SSTA.lon), layer_num * X_len])
            # print(f"{m} : {m + X_len}")
            surface_samp = np.swapaxes(surface_data[m : m + X_len, ...], 0, 3)
            layer1_samp = np.swapaxes(layer1_data[m : m + X_len, ...], 0, 3)
            layer2_samp = np.swapaxes(layer2_data[m : m + X_len, ...], 0, 3)
            
            # concat multi-layer
            t = 0
            for i in np.arange(0, X_len * layer_num, layer_num):
                X_store[..., i] = surface_samp[..., t]
                X_store[..., i+1] = layer1_samp[..., t]
                X_store[..., i+2] = layer2_samp[..., t]
                t += 1
            X_.append(X_store)
            
            slice_label = label[ m + X_len : m + X_len + y_len]
            # print(f"{m + X_len} : {m + X_len + y_len}")
            y_.append(slice_label)
        
        deal_num += 1
        print(f'No.{deal_num} model: {model_name} processed.\n')
        if deal_num == 1:
            print(f'For saving space, stop processing, total {deal_num} model.')
            break
            
    # ========================= Merge data =========================
    X = np.concatenate(X_, axis = 0)
    y = np.stack(y_, axis = 0)
    # ========================= split data ========================= 
    if X_len == 3:
        season = ['JFM', 'FMA', 'MAM', 'AMJ', 
                  'MJJ', 'JJA', 'JAS', 'ASO', 
                  'SON', 'OND', 'NDJ', 'DJF']
    if X_len == 1:
        season = ['Jan', 'Feb', 'Mar', 'Apr',
                  'May', 'Jun', 'Jul', 'Aug',
                  'Sep', 'Oct', 'Nov', 'Dec']
    encoding = nc_encoding()
    time_log = time.strftime("%Y-%m-%d_%H.%M")
    file_time_log = time.strftime("%m-%d")
    # pl.Path(f'../../temp_data/CHN_sea_forecast/{file_time_log}')\
    #   .mkdir(parents = True, exist_ok=True)
    
    if split_type == 'ALL':
        
        # create netCDF data
        sample_num = np.arange(X.shape[0], dtype = 'int32')
        channel = np.arange(X.shape[-1], dtype = 'int32')
        lead_time = np.arange(1, y.shape[-1] + 1, dtype = 'int32')
        lat, lon = SSTA.lat, SSTA.lon

        # set a new xarray Dataset contains train_data:'X', train_label:'y'
        dataset = xr.Dataset(data_vars = {
                                 'X':(['sample_num', 'lat', 'lon', 'channel'], X),
                                 'y':(['sample_num', 'lead_time'], y)
                                 },
                             coords = {
                                 'sample_num': sample_num,
                                 'lat': lat,
                                 'lon': lon,
                                 'channel': channel,
                                 'lead_time': lead_time
                                 })

        # set nc data's attributions
        dataset.attrs['data type'] = data_type
        dataset.attrs['create time'] = f'{time_log} created'
        dataset.attrs['filter'] = f"{filter_type}: {order}-order, Wn = {Wn}, signal: {filter_time} yr"
        dataset.attrs['channel'] = f'{layer_num} layers * {X_len} month'
        dataset.attrs['data source'] = f'{X_name} -> X, 10 yr after {y_name} -> y'
        dataset.attrs['model list'] = model_list
        print(dataset)
```
